```python
def format_column_name(column):
    if(column[:1] == '0'):
        mapping = {
            '01': 'gold',
            '02': 'silver',
            '03': 'bronze'
        }
        column = mapping[column[:2]] + column[4:]
    elif(column[:1] == '№'):
        column = column[2:] + '-count'
    
    return column.lower() \
                 .replace(' ', '-')

olympics.columns = map(format_column_name, olympics.columns)
```

```python
purchase_1 = pd.Series({'Name': 'Chris',
                        'Item Purchased': 'Dog Food',
                        'Cost': 22.50})
purchase_2 = pd.Series({'Name': 'Kevyn',
                        'Item Purchased': 'Kitty Litter',
                        'Cost': 2.50})
purchase_3 = pd.Series({'Name': 'Vinod',
                        'Item Purchased': 'Bird Seed',
                        'Cost': 5.00})

df = pd.DataFrame([purchase_1, purchase_2, purchase_3], index=['Store 1', 'Store 1', 'Store 2'])
```

```python
df['Location'] = df.index
df = df.set_index(['Location', 'Name'])
```

```python
df.head()
```

```python
df.loc[ [('Store 1', 'Chris'), ('Store 2', 'Vinod')] ]
```

```python
df = df.append(
    pd.Series(data={
        'Cost': 3.00,
        'Item Purchased': 'Kitty Food'
    }, name=('Store 2', 'Kevyn'))
)
```

```python
# Query the dataframe to return Kevyns stuff
kev = df[df.index.get_level_values('Name') == 'Kevyn']
kev['Cost'].mean()
```
